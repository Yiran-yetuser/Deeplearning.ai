# Broadcasting

```python
import numpy as np

A = np.array([[56.0, 0.0, 4.4, 68.0], 
            [1.2, 104.0, 52.0, 8.0], 
            [1.8, 135.0, 99.0, 0.9]])
```

```python
print(A)
```

```python
cal = A.sum(axis = 0)
print(cal)
```

axis = 0 意思是垂直求和, axis = 1 意思是水平求和。

```python
percentage = 100 * A / (cal.reshape(1, 4))  # python广播机制 (broadcasting)
print(percentage)
```

# A note on python/numpy vectors

不应该采取形如 (5, ) 来构造向量，而应该使用 (5,1)  
错误做法：

```python
a = np.random.randn(5)  # rand生成均匀分布的随机数，randn生成正态分布的随机数
print(a)
```

```python
print(a.shape)
```

这就是被称为"rand one array"的特殊数据结构  
它的行为既不像行向量也不像列向量  
DO NOT USE IT

```python
print(a.T)
```

```python
print(a.T.shape)
```

```python
print(np.dot(a,a.T))
```

正确用法如下：

```python
b = np.random.randn(5,1)
print(b)
```

```python
print(b.shape)
```

```python
print(b.T)
```

```python
print(np.dot(b,b.T))
```

可以使用“断言”语句确保不出现rand one array

```python
assert(a.shape == (5,1))
```

```python
assert(b.shape == (5,1))
```
